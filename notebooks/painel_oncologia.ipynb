```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
import os
print(os.getcwd())  # Diretório atual
```

```python
file_path = "../data/painel_oncologia.csv"  # Ajuste o caminho se necessário

# Carregar os dados
try:
    df = pd.read_csv(file_path, encoding="utf-8")
    print("Dados carregados com sucesso!")
    print(df.head())  # Exibir as primeiras linhas
except FileNotFoundError:
    print("Erro: Arquivo não encontrado. Verifique o caminho.")
```

```python
print(f"Número de linhas: {df.shape[0]}")
print(f"Número de colunas: {df.shape[1]}")
```

```python
print(df.columns)
```

```python
# Somar os casos por região
regioes = ["1 Regiao Norte", "2 Regiao Nordeste", "3 Regiao Sudeste", "4 Regiao Sul", "5 Regiao Centro-Oeste"]
df_regioes = df.loc[df["UF da residencia"] != "Total", regioes].sum()

# Criar o gráfico
plt.figure(figsize=(10, 6))
df_regioes.plot(kind="bar", color="skyblue")
plt.title("Total de Casos por Região")
plt.xlabel("Região")
plt.ylabel("Número de Casos")
plt.xticks(rotation=45)
plt.show()
```
